# **DATA CLEANING**

Pada tahap ini dilakukan proses pembersihan data untuk meningkatkan kualitas dan konsistensi dataset sebelum analisis. Proses meliputi penghapusan kolom yang tidak diperlukan, penyesuaian tipe data, standarisasi data pada beberapa variabel, serta agregasi data penjualan harian untuk menghasilkan struktur data yang lebih siap untuk dianalisis.

```python
# Hapus kolom tidak terpakai
df = df.drop(columns=[col for col in df.columns if 'Unnamed' in col])
df = df.drop(columns=['Payment_Method', 'Store_Location', 'Unit Cost', 'Transaction_ID'])
```

```python
# Ubah Unit_Price menjadi rupiah
df['Unit_Price'] = df['Unit_Price'] * 1000
```

```python
# Ubah type data
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df['Revenue'] = df['Revenue'].str.replace('.', '', regex=False).astype(float)
df['Category'] = df['Category'].astype('category')

# Cek type data kembali
df.info()
```

```python
# Ganti Tolako Minuman Herbal menjadi Tolak Linu
df['Product_Name'] = df['Product_Name'].replace({
    'Tolako Minuman Herbal': 'Tolak Linu',
    'Indomilk UHT 1L': 'Indomilk UHT Full Cream 1L'
})
```

```python
# Ganti kategori Sari Roti Tawar menjadi Bakery
df['Category'] = df['Category'].cat.add_categories('Bakery')

df.loc[
    df['Product_Name'].str.contains('Sari Roti Tawar', case=False),
    'Category'
] = 'Bakery'
```

```python
# Agregasi penjualan harian
df_grouped = df.groupby(
    ['Date', 'Province', 'Product_Name'],
    as_index=False
).agg({
    'Units_Sold': 'sum',
    'Revenue': 'sum',
    'Unit_Price': 'mean',
    'Category': 'first',
})

# Reset index
df_grouped = df_grouped.reset_index(drop=True)

# Reorder kolom
df = df_grouped[[
    'Date',
    'Product_Name',
    'Category',
    'Units_Sold',
    'Unit_Price',
    'Revenue',
    'Province',
]]
```

- Kolom yang tidak relevan dan kolom kosong berhasil dihapus sehingga dataset menyisakan variabel yang mendukung analisis.
- Tipe data pada beberapa variabel sudah disesuaikan, yaitu `Date` menjadi format *datetime*, `Revenue` menjadi numerik, serta `Category` menjadi tipe `category`.
- Dilakukan juga standarisasi pada nama produk dan kategori untuk mengurangi inkonsistensi data, yang meliputi penyesuaian nama produk serta penambahan kategori `Bakery`.

# **DATA STOCK**

Tahap ini dilakukan untuk membentuk dataset stok berdasarkan data transaksi penjualan untuk menggambarkan pergerakan persediaan produk. Proses ini mencakup pembentukan variabel `Stock_In`, `Stock_Out`, dan `Stok_End`. Pengurutan data berdasarkan tanggal dan produk, serta simulasi penambahan stok secara terjadwal maupun darurat agar ketersediaan stok tetap terjaga selama periode penjualan.

```python
# Load dataset buntuk data stock
df_transactions = df
```

```python
# Membuat kolom baru untuk stock
df_transactions["Stock_Out"] = df_transactions["Units_Sold"]
df_transactions["Stock_In"] = 0
df_transactions["Stock_End"] = 0
```

```python
# Urutkan data berdasarkan Product_Name dan Date
df_transactions = df_transactions.sort_values(
    by=["Product_Name", "Date"]
).reset_index(drop=True)
```

```python
# Membuat hasil random selalu sama setiap run
random.seed(42)

# Simpan stock terakhir produk
stock_product = {}

# Hari restock toko
restock_day = ["Monday", "Thursday"]

for i in range(len(df)):
    product = df_transactions.loc[i, "Product_Name"]
    date = df.loc[i, "Date"]

    # Stok awal produk
    if product not in stock_product:
        additional_stock = random.randint(300, 700)
        df_transactions.loc[i, "Stock_In"] = additional_stock
        stock = additional_stock
    else:
        stock = stock_product[product]

    # Jumlah stock out
    stock_out = df_transactions.loc[i, "Stock_Out"]

    # Restock terjadwal
    if date.day_name() in restock_day:
        if stock < 200:
            additional_stock = random.randint(250, 600)
            df_transactions.loc[i, "Stock_In"] += additional_stock
            stock += additional_stock

    # Restock darurat
    if stock < stock_out:
        additional_stock = random.randint(300, 700)
        df_transactions.loc[i, "Stock_In"] += additional_stock
        stock += additional_stock

    # Stock berkurang karena penjualan
    stock -= stock_out

    # Batas stock
    stock = min(stock, 1000)
    stock = max(stock, 0)

    # Simpan stock end
    df_transactions.loc[i, "Stock_End"] = stock
    stock_product[product] = stock
```

```python
# Dataset akhir
df_stock = df_transactions[[
    "Date",
    "Product_Name",
    "Stock_In",
    "Stock_Out",
    "Stock_End"
]].copy()

# Urutkan berdasarkan Date dan Product_Name
df_stock = df_stock.sort_values(
    by=["Date", "Product_Name"]
).reset_index(drop=True)

# Cek isi dataset
df_stock.head()
```

```python
# Save ke CSV
df_stock.to_csv("../data/processed/transactions/stock_raw.csv", index=False)
```

**Insight:**

- Dataset stok berhasil dibentuk dengan penambahan informasi `Stock_In`, `Stock_Out`, dan `Stock_End` untuk menggambarkan pergerakan persediaan pada setiap produk.
- Penerapan simulasi restock terjadwal dan restock darurat memungkinkan stok tetap tersedia ketika jumlah persediaan menurun atau tidak mencukupi jumlah penjualan.
- Pengurutan data berdasarkan produk dan tanggal membantu menjaga konsistensi perhitungan stok dari waktu ke waktu.
- Dataset menghasilkan informasi persediaan yang lebih terstruktur sehingga dapat digunakan untuk analisis stok.


# **MERGE DATASET TRANSACTIONS DAN STOCK**

Tahap ini dilakukan untuk menggabungkan dataset transaksi dan dataset stok untuk menghasilkan data yang lebih lengkap. Proses ini mencakup penghapusan kolom yang tidak diperlukan pada dataset transaksi, penggabungan data berdasarkan `Date` dan `Product_Name`, serta pengurutan ulang dataset untuk menjagka konsistensi data pada proses analisis selanjutnya.

```python
# Hapus kolom yang tidak dipakai pada dataset transaction
df_transactions = df_transactions.drop(
    columns=[
        "Stock_In",
        "Stock_Out",
        "Stock_End"
    ]
)
```

```python
# Gabung kedua dataset
df_merge = pd.merge(
    df_transactions,
    df_stock,
    on=["Date", "Product_Name"],
    how="left"
)

# Urutkan berdasarkan Date dan Product_Name
df_merge = df_merge.sort_values(
    by=["Date", "Product_Name"]
).reset_index(drop=True)

# Cek isi dataset
df_merge.head()
```

```python
# Save ke CSV
df_merge.to_csv("../data/processed/transactions/transactions_clean.csv", index=False)
```

**Insight:**

- Dataset transaksi dan stok berhasil digabungkan sehingga menghasilkan dataset yang lebih lengkap.
- Penggabungan data berdasarkan `Date` dan `Product_Name` memungkinkan informasi transaksi dan pergerakan stok saling terhubung secara konsisten.
- Pengurutan ulang data berdasarkan tanggal dan nama produk menghasilkan dataset yang lebih rapi dan mudah digunakan pada proses analisis selanjutnya.
- Dataset akhir telah memuat informasi transaksi sekaligus stok sehingga lebih siap digunakan untuk eksplorasi data maupun analisis lanjutan.
