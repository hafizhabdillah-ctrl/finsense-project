# **MONO**


Tahap ini dilakukan untuk mengubah audio menjadi satu channel *(mono)* agar menghasilkan format audio yang lebih konsisten serta mempermudah tahap preprocessing berikutnya.

```python
os.makedirs("../data/processed/audio", exist_ok=True)

input_folder = "../data/raw/audio/audio_double_raw/audio_1200"
output_folder = "../data/processed/audio/audio_mono"

os.makedirs(output_folder, exist_ok=True)

# Ubah audio ke mono
for file in sorted(os.listdir(input_folder)):
  if file.endswith(".wav"):
    input_path = os.path.join(input_folder, file)
    try:
      audio = AudioSegment.from_wav(input_path)
      mono_audio = audio.set_channels(1)

      # Save file
      output_path = os.path.join(output_folder, file)
      mono_audio.export(output_path, format="wav")

      print(f"Berhasil: {file}")
    except Exception as e:
      print(f"Gagal: {file} > {e}")

print("\nSemua Audio Berhasil Diubah ke Mono")
```

**Insight:**

Seluruh file audio berhasil dikonversi menjadi format mono (`1 channel`) sehingga menghasilkan audio yang lebih konsisten dan siap digunakan untuk tahap preprocessing berikutnya.

# **RESAMPLING (16 kHz)**

Tahap ini dilakukan untuk mengubah dan menyeragamkan *sample rate* audio menjadi `16 kHz` agar format audio lebih konsisten dan sesuai untuk tahap berikutnya.

```python
input_folder = "../data/processed/audio/audio_mono"
output_folder = "../data/processed/audio/audio_16kHz"

os.makedirs(output_folder, exist_ok=True)

# Ubah sample rate ke 16kHz
for file in sorted(os.listdir(input_folder)):
  if file.endswith(".wav"):
    input_path = os.path.join(input_folder, file)
    try:
      audio = AudioSegment.from_wav(input_path)
      audio_16kHz = audio.set_frame_rate(16000)

      # Save file
      output_path = os.path.join(output_folder, file)
      audio_16kHz.export(output_path, format="wav")

      print(f"Berhasil: {file}")
    except Exception as e:
      print(f"Gagal: {file} > {e}")

print("\nSemua Audio Berhasil Diubah ke 16kHz")
```

**Insight:**

Seluruh audio berhasil diseragamkan ke *sample rate* `16 kHz` sehingga variasi sampling antar audio dapat diminimalkan dan format audio menjadi lebih konsisten untuk tahap berikutnya.

# **NORMALIZE**

Tahap ini dilakukan untuk menstabilkan tingkat volume audio agar intensitas suara antar audio menjadi lebih konsisten dan siap digunakan pada tahap berikutnya.

```python
input_folder = "../data/processed/audio/audio_16kHz"
output_folder = "../data/processed/audio/audio_normalized"

os.makedirs(output_folder, exist_ok=True)

# Normalize volume
for file in sorted(os.listdir(input_folder)):
  if file.endswith(".wav"):
    input_path = os.path.join(input_folder, file)
    try:
      audio = AudioSegment.from_wav(input_path)
      normalized_audio = normalize(audio)

      # Save file
      output_path = os.path.join(output_folder, file)
      normalized_audio.export(output_path, format="wav")

      print(f"Berhasil: {file}")
    except Exception as e:
      print(f"Gagal: {file} > {e}")

print("\nSemua Audio Berhasil Dinormalisasi")
```

**Insight:**

Proses normalisasi berhasil dilakukan untuk menstabilkan tingkat volume audio sehingga perbedaan *loudness* antar audio menjadi lebih konsisten dan audio siap digunakan pada tahap berikutnya.

# **TRIM SILENCE**

Tahap ini dilakukan untuk menghapus bagian hening *(silence)* pada awal dan akhir audio agar audio lebih fokus pada suara utama serta mengurangi bagian yang tidak diperlukan.

```python
input_folder = "../data/processed/audio/audio_normalized"
output_folder = "../data/processed/audio/audio_trimmed"

os.makedirs(output_folder, exist_ok=True)

# Trim Silence
for file in sorted(os.listdir(input_folder)):
  if file.endswith(".wav"):
    input_path = os.path.join(input_folder, file)
    try:
      audio = AudioSegment.from_wav(input_path)

      # Deteksi bagian non-hening
      non_silence = detect_nonsilent(
          audio,
          min_silence_len=500,
          silence_thresh=-40
      )

      # Jika audio mengandung bagian non-hening
      if non_silence:
        start = non_silence[0][0]
        end = non_silence[-1][1]

        trimmed_audio = audio[start:end]
      else:
        trimmed_audio = audio

      # Save file
      output_path = os.path.join(output_folder, file)
      trimmed_audio.export(output_path, format="wav")

      print(f"Berhasil: {file}")
    except Exception as e:
      print(f"Gagal: {file} > {e}")

print("\nSemua Audio Berhasil Melalui Trim Silence")
```

**Insight:**

Bagian hening pada awal dan akhir audio berhasil dikurangi sehingga audio menjadi lebih fokus pada suara utama dan mengurangi bagian tidak relevan.
