# **A/B TESTING**

Pada FinSense, pengguna dapat melakukan pencatatan transaksi melalui pengetikan *(manual input)* maupun melalui fitur *(voice input)*. Metode *manual input* mengharuskan pengguna mengetik informasi transaksi secara langsung, sedangkan *voice input* memungkinkan pengguna mencatat transaksi melalui suara.

Pengujian ini bertujuan untuk mengetahui apakah fitur *voice input* pada FinSense mampu meningkatkan efisiensi waktu pencatatan transaksi dibandingkan metode manual. Evaluasi dilakukan dengan membandingkan waktu yang dibutuhkan pengguna untuk menyelesaikan pencatatan transaksi pada kedua metode.

Metrik utama yang digunakan dalam pengujian ini adalah `recording_time`, yaitu waktu yang dibutuhkan pengguna untuk menyelesaikan proses pencatatan transaksi.

Pengujian dilakukan pada dua kelompok berikut:

* **Manual Input**: Kelompok yang melakukan pencatatan transaksi melalui pengetikan.
* **Voice Input**: Kelompok yang melakukan pencatatan transaksi menggunakan fitur suara.


## Hipotesis

Hipotesis yang digunakan dalam pengujian ini adalah sebagai berikut:

- $H_{0}$ (Hipotesis nol): Rata-rata waktu pencatatan transaksi menggunakan *voice input* tidak lebih cepat dibandingkan *manual input*.

- $H_{1}$ (Hipotesis alternatif): Rata-rata waktu pencatatan transaksi menggunakan *voice input* lebih cepat dibandingkan *manual input*.

## Import Library

Tahap ini dilakukan untuk mengimpor berbagai library yang akan digunakan dalam proses pengujian.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

## Menyiapkan Data

Pada tahap ini, data waktu transaksi hasil pengujian disiapkan ke dalam dua kelompok, yaitu *manual input* dan *voice input*. Masing-masing kelompok berisi 20 percobaan dengan nilai berupa durasi transaksi dalam satuan waktu. Data tersebut akan digunakan untuk membandingkan efisiensi waktu transaksi antara kedua metode melalui pengujian A/B testing.


```python
# Memisahkan data berdasarkan metode pencatatan
manual_time = [
    29.69, 24.87, 26.28, 24.32, 25.76,
    24.50, 22.08, 22.33, 25.63, 24.58,
    23.52, 25.80, 27.50, 24.52, 26.52,
    25.08, 25.48, 22.34, 24.97, 21.08
]

voice_time = [
    25.71, 23.62, 22.15, 22.31, 22.89,
    22.38, 23.76, 22.15, 23.37, 23.12,
    23.86, 23.61, 23.87, 24.82, 23.61,
    22.70, 22.22, 22.15, 24.06, 22.23
]

print(f"Jumlah data manual: {len(manual_time)}")
print(f"Jumlah data voice: {len(voice_time)}")
```

**Insight:**

Data terdiri dari dua kelompok, yaitu manual input dan voice input, yang masing-masing memiliki 20 sampel waktu transaksi. Jumlah sampel yang seimbang pada kedua kelompok membantu proses perbandingan dilakukan secara lebih konsisten.

## Statistik Deskriptif

Statistik deskriptif digunakan untuk melihat gambaran awal waktu pencatatan transaksi pada masing-masing metode.

```python
manual_df = pd.Series(manual_time)
voice_df = pd.Series(voice_time)

print("Manual Input")
print(manual_df.describe())

print("\nVoice Input")
print(voice_df.describe())
```

**Insight:**

- Metode *voice input* memiliki rata-rata waktu transaksi sebesar 23,23 detik, lebih rendah dibandingkan *manual input* yang memiliki rata-rata 24,84 detik. Hal ini menunjukkan bahwa *voice input* cenderung menyelesaikan transaksi lebih cepat.
- Nilai median pada *voice input* sebesar 23,25 detik juga lebih rendah dibandingkan *manual input* yang memiliki median sebesar 24,92 detik. Hal ini mengindikasikan bahwa sebagian besar transaksi dengan *voice input* berlangsung lebih singkat.
- Standar deviasi *voice input* sebesar 0,98 detik lebih kecil dibandingkan *manual input* yang memiliki standar deviasi sebesar 1,99 detik. Hal ini menunjukkan bahwa waktu transaksi pada *voice input* lebih konsisten dan memiliki variasi yang lebih rendah.
- Rentang waktu transaksi *voice input* berada pada 22,15–25,71 detik, sedangkan *manual input* berada pada 21,08–29,69 detik. Hal ini menunjukkan bahwa metode *manual input* memiliki variasi waktu yang lebih besar.


## Uji Hipotesis

Pengujian hipotesis dilakukan menggunakan *independent samples t-test* (uji t dua sampel independen) untuk membandingkan rata-rata waktu pencatatan transaksi antara metode *manual input* dan *voice input*. Karena pengujian bertujuan mengetahui apakah *voice input* lebih cepat dibandingkan *manual input*, maka digunakan pengujian satu sisi (*one-tailed test*).

```python
# Uji t
t_statistic, p_value = stats.ttest_ind(
    voice_time,
    manual_time,
    alternative="less"
)

print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

**Insight:**

### Insight Uji Hipotesis

- Hasil independent sample t-test menunjukkan nilai p-value sebesar 0,0012 (< 0,05), sehingga $H_{0}$ ditolak. Hal ini berarti bahwa terdapat perbedaan yang signifikan secara statistik antara rata-rata waktu transaksi metode *manual input* dan *voice input*.
- Nilai t-statistic yang negatif (-3,25) menunjukkan bahwa rata-rata waktu transaksi *voice input* lebih rendah dibandingkan *manual input*. Dengan demikian dapat disimpulkan bahwa penggunaan *voice input* secara signifikan lebih cepat dalam menyelesaikan transaksi dibandingkan metode *manual input*.

## Interpretasi Hasil

Interpretasi dilakukan berdasarkan nilai p-value dengan tingkat signifikansi sebesar 0,05.

- Jika p-value < 0,05, maka $H_{0}$ ditolak. Hal ini menunjukkan bahwa terdapat bukti yang cukup untuk menyimpulkan bahwa voice input pada FinSense lebih cepat dibandingkan manual input dalam pencatatan transaksi.
- Jika p-value ≥ 0,05, maka $H_{0}$ gagal ditolak. Hal ini menunjukkan bahwa belum terdapat bukti yang cukup untuk menyimpulkan bahwa voice input lebih cepat dibandingkan manual input dalam pencatatan transaksi.

```python
alpha = 0.05

if p_value < alpha:
    print("H0 ditolak")
    print("Voice input lebih efisien dibanding manual input")
else:
    print("H0 gagal ditolak")
    print("Belum ada bukti bahwa voice input lebih efisien")
```

**Insight:**

Hasil pengujian menunjukkan bahwa nilai p-value yang diperoleh lebih kecil dari 0,05, sehingga $H_{0}$ ditolak. Hal ini menunjukkan bahwa terdapat perbedaan yang signifikan secara statistik antara metode *voice input* dan *manual input* dalam proses pencatatan transaksi.

Lebih lanjut, hasil ini mengindikasikan bahwa penggunaan *voice input* pada FinSense memberikan efisiensi waktu yang lebih baik dibandingkan metode manual. Dengan demikian, penggunaan *voice input* dapat menjadi alternatif yang lebih efektif dalam meningkatkan kecepatan pencatatan transaksi oleh pengguna.

## Kesimpulan

Berdasarkan hasil pengujian, diperoleh nilai *p-value* < 0,05 sehingga $H_{0}$ ditolak. Hasil ini menunjukkan bahwa *voice input* pada FinSense secara signifikan lebih cepat dibandingkan *manual input* dalam proses pencatatan transaksi, sehingga dapat meningkatkan efisiensi waktu pencatatan.
